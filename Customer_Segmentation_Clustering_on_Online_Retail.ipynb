# **Segmentation / Clustering using K-means**

# **Principal Component Analysis (PCA)**

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
points = pca.fit_transform(df[cols])
```

```python
df['PC_1'] = points[:,0]
df['PC_2'] = points[:,1]
```

```python
df
```

```python
sum(pca.explained_variance_ratio_)*100
```

```python
plt.figure(1 , figsize = (10,10))
sns.scatterplot(x='PC_1', y='PC_2', data=df, s=50)
```

# **The Elbow Method**

```python
# Let's find right fit for K
from sklearn.cluster import KMeans
import numpy as np

# Within Cluster Sum of Squared Errors (WCSS) 
wcss = []
max_range = 16
for i in range(1, max_range):
    km = KMeans(n_clusters = i, init = 'k-means++')
    km.fit(df[cols])
    wcss.append(km.inertia_)
    
plt.figure(1 , figsize = (7, 4))
plt.plot(np.arange(1 , max_range) , wcss , 'o')
plt.plot(np.arange(1 , max_range) , wcss , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS')
plt.show()
```

**Here we see the Elbow Method  got the last breakage exactly at 3. After 3 it has a continuous line. So, the no. of cluster should be 3.**

# **K-Means Algorithm** 

```python
algorithm = KMeans(n_clusters = 3) 
algorithm.fit(df[cols])
grp = algorithm.labels_
df['group'] = grp
df
```

#**Analysing the 3 clusters with the help of Scatter Plot.**

```python
plt.figure(1 , figsize = (8, 6))
sns.scatterplot(x='PC_1', y='PC_2', hue='group', data=df, s=100, palette="Set1")
```

**Here we see 3 types of clusters.**

```python
grp0 = df[df["group"]==0].copy()
grp1 = df[df["group"]==1].copy()
grp2 = df[df["group"]==2].copy()
```
